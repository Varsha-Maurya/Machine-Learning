# B. Probability Theory

## Probability Theory

In real life scenarios, Probability theory provides simple solutions even though Not completely accurate solutions. This type of learning is extensively used in Machine learning.

##### Random Variable: A random variable is nothing but a result of any random action. 
Like in case of a coin toss (Random Action) you might get a Heads or Tails (Random output).

##### Frequentist probability: Frequentist probability as the name suggests, takes into consideration the frequency of the any event happening out of large number of same trials. 

##### Probability distribution: It is a representation of how likely a random variable might take a specific value out of all its possible values. 
E.g. : A dice getting Number 1 when rolled.

##### Marginal Probability: This is the probability of any single event irrespective of any other event. 
Ex.: If we draw a single card from a deck of cards, the probability of it being a 2 is 1/13 (4/52).

##### Conditional Probability: This is the probability of any event occurring only if another event also takes place.
Ex.: The probability of us drawing 2 from a deck of card if it’s a red card is 1/13 (2/26).

##### The Chain Rule of Conditional Probabilities : The probability of 2 or more events happening one after the other can be calculated using this rule. 
Ex.: If we want to know the probability of selecting a 1 card (8 of any suit) out of 2 decks available then the probability can be calculated as – 
1 /2 (Probability of selecting 1 deck of card) * 4/52 (Probability of selection 1 card out of the selected deck)

##### Independence and Conditional Independence: 2 or more events can be called independent if outcome of the 1st event does not have any impact on the outcome of 2nd event.
In case of Conditional Independence, while the 2 events are independent from each other but they might be dependent on some other event which makes them only conditionally Independent.
Ex.: If we say we want to know the probability of drawing 2 cards (Both Number 9) from a deck of cards 1 by 1 (Assuming we placed the card back into the deck after 1st draw). Then these events are Independent events. But if we say that the Sum of both the cards is even then its conditional dependency.

##### Expectation, Variance and Covariance: Expectation is basically the mean value of the possible outcome of some trials.
Variance is the deviation in the value of a variable from the given mean value for the trial.

Covariance: 
Ex.: Expected value of a card being drawn from a deck of card is 7. If we draw a 5 numbered card then the variance can be depicted as 2.

##### Bayesian probability
Bayesian probability is an interpretation of the concept of probability, in which, instead of frequency or propensity of some phenomenon, probability is interpreted as reasonable expectation representing a state of knowledge or as quantification of a personal belief


##### Discrete Variables and Probability Mass Functions
The probability mass function maps from a state of a random variable tothe probability of that random variable taking on that state. The probabilitythatx=xis denoted asP(x), with a probability of 1 indicating thatx=xiscertain and a probability of 0 indicating thatx=xis impossible.

## Pre processing of the data
1. Data pre processing. Since there are NaN values in some of the columns we will remove it.
2. We will calculate the total number of NaN values for each column in the dataset 
2. Since every columns has something important we will drop the column that has the highest NaN and see if the total rows doesn't drop considerbly.
3. If the rows doesn't drop considerably and there are no more NaN we can use that as our dataset.
4. If the rows drop considerably to the original size or there are still considerable about of NaN then we will perform data-imputation on the dataset.

#### Get the count of NaN for all columns and see which one has the highest NaN'.s

```python

census_raw.isnull().sum()
```

#### Since ChildPoverty column has the highest number of NaN's.We will remove all the rows where ChildPoverty==NaN

```python
census_raw = census_raw[pd.notnull(census_raw['ChildPoverty'])]
# Get the count of NaN for all columns.
census_raw.isnull().sum()
```

#### We will perform a simple data imputation technique now to have all the values. Calculate the median and mean of the columns and fill it using pandas. 
    1. Income colums are usually more closely associated with median 
    2. Other counts like people in different sectors are closely associated with mean.


#### Median

```python
colMed = ["Income","IncomeErr","IncomePerCap","IncomePerCapErr"]
median = census_raw[colMed].median(axis=0) #along columns.
median
```

#### Mean

```python
colMean = ["Professional","Service","Office","Construction","Production",
       "Drive","Carpool","Transit","Walk","OtherTransp","WorkAtHome",
       "MeanCommute","PrivateWork","PublicWork","SelfEmployed",
       "FamilyWork","Unemployment"]
mean = census_raw[colMean].mean(axis=0) #along columns.
mean
```

#### Fill median to the dataset of missing values in it's columns.

```python
census_raw.fillna(median,inplace=True);
```

#### Fill mean to the dataset of missing values in it's columns.

```python
census_raw.fillna(mean,inplace=True);
```

#### Verify there are no more NaN's in the DF.

```python
census_raw.isnull().sum()
```

#### Data Transformation

```python
StateIncome = census_raw.groupby('State').apply(lambda x: x['Income'].sum()/len(x))
StateEthinicity = census_raw.groupby('State').apply(lambda x: x[['Hispanic','White','Black','Native','Asian','Pacific']].sum()/len(x))
StateGender = census_raw.groupby('State').apply(lambda x: x[['Men','Women']].sum()/len(x))
StateGenderPer = pd.concat([StateGender['Men']*100/StateGender.sum(1), StateGender['Women']*100/StateGender.sum(1)], axis=1,keys=['Men','Women'])
StateWorkDistribution = census_raw.groupby('State').apply(lambda x: x[['PrivateWork','PublicWork','SelfEmployed','FamilyWork','Unemployment']].sum()/len(x))
StateWorkingCategory = census_raw.groupby('State').apply(lambda x: x[['Professional','Service','Office','Construction','Production']].sum()/len(x))

StatePoverty = census_raw.groupby('State').apply(lambda x: x[['Poverty']].sum()/len(x))
StateChildPoverty = census_raw.groupby('State').apply(lambda x: x[['Poverty','ChildPoverty']].sum()/len(x))
WorkCommute = census_raw.groupby('State').apply(lambda x:x[['Drive','Carpool','Transit','Walk','OtherTransp','WorkAtHome']].sum()/len(x))
MeanCommute = census_raw.groupby('State').apply(lambda x:x[['MeanCommute']].sum()/len(x))

TotalPop = census_raw.groupby('State').apply(lambda x:x[['TotalPop']].sum())
```
